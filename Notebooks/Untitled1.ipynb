```python
class Solution:
    def partitionLabels(self, S: str) -> List[int]:
        ## RC ##
        ## APPROACH : GREEDY ##
        ## LOGIC ##
        ## 1. We store First and last index of all alphabets
        ## 2. We start with one window i.e 0 to 8 from below example, and we go on with next chars
        ## 3. If both start and end of next char fall in between current window, then donot append to result
        ## 4. If start of next char is in current window and end is not, then we extend the current window
        ## 5. If start itself is not present in the current window, we create new window
        
        ## EXAMPLE ## "ababcbacadefegdehijhklij"
        ## STACK TRACE ##
        ## defaultdict(<class 'list'>, {'a': [0, 2, 6, 8], 'b': [1, 3, 5], 'c': [4, 7], 'd': [9, 14], 'e': [10, 12, 15], 'f': [11], 'g': [13], 'h': [16, 19], 'i': [17, 22], 'j': [18, 23], 'k': [20], 'l': [21]})
        # [] 0 8
        # [0, 8] 1 5
        # [0, 8] 4 7
        # [0, 8] 9 14
        # [9, 14] 10 15
        # [9, 15] 11 11
        # [9, 15] 13 13
        # [9, 15] 16 19
        # [16, 19] 17 22
        # [16, 22] 18 23
        # [16, 23] 20 20
        # [16, 23] 21 21


        
        charIndex = collections.defaultdict(list)
        for i, ch in enumerate(S):
            charIndex[ch].append(i)
        # print(charIndex)
        
        windows = []
        window = []
        for ch in list(charIndex):
            
            # print(window, charIndex[ch][0], charIndex[ch][-1])
            if( not window ):
                window = [charIndex[ch][0], charIndex[ch][-1]]
                continue
            # 3
            if( window[0] <= charIndex[ch][0] <= charIndex[ch][-1] <= window[1] ):
                continue
            
            # 4
            if( window[0] <= charIndex[ch][0] <= window[1] ):
                window[1] = charIndex[ch][-1]
                continue
                
            # 5
            if( charIndex[ch][0] > window[0] ):
                windows.append(window)
                window = [charIndex[ch][0], charIndex[ch][-1]]
        
        #append the last window
        windows.append(window)
        return [ w[1]-w[0]+1 for w in windows]
```
